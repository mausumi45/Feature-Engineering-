### Feature enginerring

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month'] = df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['year'] = df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
### Drop date of journey column
df.drop('Date_of_Journey',axis = 1, inplace = True)
```

```python
df
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
df.drop('Arrival_Time', axis =1, inplace = True)
```

```python
df
```

```python
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
df.drop('Dep_Time', axis =1, inplace = True)
```

```python
df
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, np.nan : 1, '4 stops':4})
```

```python
df.info()
```

```python
df['Duration_hour'] = df['Destination'].str.split(' ').str[0].str.split('h').str[0].astype(int)
df['Duration_min'] =  df['Destination'].str.split(' ').str[1].str.split('h').str[0].astype(int)
```
